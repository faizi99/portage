```python
form_array=[]
table_array=[]
```

```python
def common_data(list1, list2):
    result = False
  
    # traverse in the 1st list
    for x in list1:
  
        # traverse in the 2nd list
        for y in list2:
    
            # if one common
            if x == y:
                result = True
                return result 
                  
    return result
```

```python
def process_text_analysis(bucket, document):
    
    global key_value_words_array
    global table_cell_words_array
    global table_array
    global table_counter
    global key_value_set_counter

    #Get the document from S3
    
                          
    s3_object = s3_connection.Object(bucket,document)
    s3_response = s3_object.get()

    stream = io.BytesIO(s3_response['Body'].read())
    image=Image.open(stream)

    # Analyze the document
    #new_client = boto3.client('textract')
    
    image_binary = stream.getvalue()
    new_response = client.analyze_document(Document={'Bytes': image_binary},
        FeatureTypes=["TABLES", "FORMS"])

    ### Alternatively, process using S3 object ###
    #response = client.analyze_document(
    #    Document={'S3Object': {'Bucket': bucket, 'Name': document}},
    #    FeatureTypes=["TABLES", "FORMS"])

    ### To use a local file ###
    # with open("pathToFile", 'rb') as img_file:
        ### To display image using PIL ###
    #    image = Image.open()
        ### Read bytes ###
    #    img_bytes = img_file.read()
    #    response = client.analyze_document(Document={'Bytes': img_bytes}, FeatureTypes=["TABLES", "FORMS"])

    
    #Get the text blocks
    blocks=new_response['Blocks']
    #width, height =image.size  
    #draw = ImageDraw.Draw(image)  
    #print ('Detected Document Text')
    
   
    # Create image showing bounding box/polygon the detected lines/text
            
    for block in blocks:
        if block['BlockType'] == "KEY_VALUE_SET":
            key_value_set_counter+=1
            try:
                for relation_element in block['Relationships']:
                    if relation_element['Type'] == 'CHILD':
                        for unique_identity in relation_element['Ids']:
                            key_value_words_array.append(unique_identity)
                            #print(unique_identity)
            except KeyError:
                pass
    #print(key_value_words_array)
                
                
                
    for block in blocks:        
        if block['BlockType'] == "TABLE":
            for relation_element in block['Relationships']:
                if relation_element['Type'] == 'CHILD':
                    my_cell_counter= len(relation_element['Ids'])
                    for unique_identity in relation_element['Ids']:
                        matcher= unique_identity
                        for block in blocks:
                            if block['Id'] == matcher:
                                try:
                                    for relation_element in block['Relationships']:
                                        if relation_element['Type'] == 'CHILD':
                                            for unique_identity in relation_element['Ids']:
                                                table_cell_words_array.append(unique_identity)
                                except KeyError:
                                    pass
            #print(table_cell_words_array)
            duplicate= common_data(key_value_words_array, table_cell_words_array)                            
            if duplicate  == True and key_value_set_counter > 15:
                    form_array.append(document)
            if duplicate == False and my_cell_counter >15:
                print("this is a table")
                table_counter+=1
                table_array.append(document)
                
    if table_counter ==0 and key_value_set_counter >15:
            form_array.append(document)
            
    if table_counter > 0 and key_value_set_counter >15:
            form_array.append(document)
        
            
        
        
        

            
        #if block['BlockType'] == 'TABLE':
        #table_counter = table_counter+1
      

            
            
                
       
   
            #uncomment to draw polygon for all Blocks
            #points=[]
            #for polygon in block['Geometry']['Polygon']:
            #    points.append((width * polygon['X'], height * polygon['Y']))
            #draw.polygon((points), outline='blue')
            
    #return (table_counter)


```

```python
def table_count(file_name):
    bucket = 'johnsplit'
    document = file_name
    process_text_analysis(bucket,document)
```

```python
container_folder="container"
folder_id=0
csv_number=0
for folder in os.listdir(container_folder):
    #our important variables
    key_value_set_counter=0
    key_value_words_array=[]
    table_cell_words_array=[]
    multi_question_mark=0
    multi_dollar_sign=0
    table_counter=0
    test_counter=0
    date_counter=0
    medication_counter=0
    icd_counter=0
    
    
    folder_id+=1
    path_folder=f"./container/folder{folder_id}"
    for i in range(len(os.listdir(path_folder))):
        with open(f"./container/folder{folder_id}/doc{folder_id}_split{i}.jpg", 'rb') as image:
            image_bytes = image.read()
        textract_response = client.detect_document_text(Document={'Bytes': image_bytes})
        ourtext = ""
        for item in textract_response["Blocks"]:
            if item["BlockType"] == "LINE":
                #print ('\033[94m' +  item["Text"] + '\033[0m')
                ourtext = ourtext + " " + item["Text"]
        multi_question_mark+= ourtext.count('?')
        multi_dollar_sign+= ourtext.count('$')
        table_count(f"doc{folder_id}_split{i}.jpg")
        #print("question mark: " + str(multi_question_mark))
        #print("dollar sign: " + str(multi_dollar_sign))
        #print("tables: " + str(table_counter))
        entities = new_client.detect_entities(Text=ourtext)
        for entity in entities['Entities']:
            if entity["Type"] == "TEST_NAME":
                test_counter+=1
            if entity["Type"] == "DATE":
                date_counter+=1
            if entity["Category"] =="MEDICATION":
                medication_counter+=1
        icd_entities = new_client.infer_icd10_cm(Text=ourtext)
        for icd_entity in icd_entities['Entities']:
            if (icd_entity['ICD10CMConcepts'] != []):
                icd_counter+=1
                
                
                
    #making csv
    
    with open(f'out{csv_number+1}.csv', 'w',newline="") as f: # output csv file
        writer = csv.writer(f)
        with open(f'out{csv_number}.csv','r') as csvfile: # input csv file
            reader = csv.reader(csvfile, delimiter=',')
            row_number=0
            for row in reader:
                if row_number==csv_number+1:
                    row[15] = table_counter
                    row[16] = multi_question_mark
                    row[17] = multi_dollar_sign
                    row[18] = test_counter
                    row[19] = icd_counter
                    row[20] = medication_counter
                    row[21] = date_counter

                    writer.writerow(row)
                else:
                    writer.writerow(row)
                row_number+=1
            csv_number+=1
```
